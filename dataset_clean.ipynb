```python
import pandas as pd

# Load dataset
df = pd.read_csv("D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Sales Performance & Revenue Growth Dashboard/Sales_Raw Data.csv")

# Remove missing values
df = df.dropna()

# Rename columns (example if needed)
df.columns = df.columns.str.strip()

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
# Remove invalid dates
df = df.dropna(subset=['Date'])

# Convert numeric columns
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Price'] = pd.to_numeric(df['Price'])

# Create Sales Amount column
df['Sales Amount'] = df['Quantity'] * df['Price']

# Save cleaned dataset
df.to_csv("D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Sales Performance & Revenue Growth Dashboard/Sales_cleaned.csv", index=False)

print("Cleaning Done. File saved as Sales_cleaned.csv")
```
